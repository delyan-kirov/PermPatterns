# Scripts

## Vincular Mesh

```python
def genVincMesh(length:int, pattern:List[int], indeces:List[int]) -> List[int]:
    points = []
    for i in indeces: points += [(i,x) for x in range(0,length)]
    VS.visualize(pattern, points)
    return points

genVincMesh(4, [1,3,2], [1,3])
```

## Bivincular Mesh

```python
def genBivincMesh(length:int, pattern:List[int], indeces:Tuple[List[int], List[int]]) -> List[int]:
    points = []
    for i in indeces[0]: points += [(i,x) for x in range(0,length)]
    for i in indeces[1]: points += [(x,i) for x in range(0,length)]
    VS.visualize(pattern, points)
    return points

genBivincMesh(4, [1,3,2], ([1], [2]))
```

## Consecutive Mesh

```python
def genConsecMesh(length:int, pattern:List[int]) -> List[int]:
    indeces = range(min(pattern), max(pattern))
    points = []
    for i in indeces: points += [(i,x) for x in range(0,length)]
    VS.visualize(pattern, points)
    return points

genConsecMesh(4, [1,3,2])
```

## Boxed Mesh

```python
# The length doesn't matter but let's keep it anyway. 
def genBoxMesh(length:int, pattern:List[int]) -> List[int]:
    indeces = range(min(pattern), max(pattern))
    rows = []
    cols = []
    for i in indeces: rows += [(x,i) for x in range(0,length)]
    for i in indeces: cols += [(i,x) for x in range(0,length)]
    points = [point for point in rows if point in cols]
    VS.visualize(pattern, points)
    return points

genBoxMesh(4, [1,3,2])
```
